```python
#Import Dependecies
import pandas as pd
```

```python
# Files to Load

vine_to_load = "vine_table.csv"
```

```python
# Read the file and save it to a Pandas DataFrame

vine_data_df = pd.read_csv(vine_to_load)
vine_data_df.head(10)
```

```python
#Filter the data to retrieve all of the rows where the total votes count is greater than or equal to 20
filter1_df = vine_data_df[vine_data_df['total_votes']>=20]
filter1_df.head(10)
```

```python
#Filter the new DataFrame and create a new DataFrame to retrieve all of the rows where
#the number of helpful votes divided by total votes if greater than or equal to 50%
filter2_df = filter1_df[(filter1_df['helpful_votes']/filter1_df['total_votes'])>=0.5]
filter2_df.head(10)
```

```python
#Filter the DataFrame and create a new DataFrame that retrieves all of the rows
#where a review was written as part of the Vine program (paid)
vine_paid_df = filter2_df[filter2_df['vine'] == 'Y']
vine_paid_df.head(10)
```

```python
#Filter the filter2 DataFrame and create a new DataFrame that retrievs all of the rows
#where a review was not part of the Vine program (unpaid)
vine_unpaid_df = filter2_df[filter2_df['vine'] == 'N']
vine_unpaid_df.head(10)
```

```python
#Total number of unpaid reviews
len(vine_unpaid_df)
```

```python
#Total number of unpaid 5-star reviews
len(vine_unpaid_df[vine_unpaid_df['star_rating'] == 5.0])
```

```python
#Percentable of unpaid 5-star reviews
len(vine_unpaid_df[vine_unpaid_df['star_rating'] == 5.0])/len(vine_unpaid_df)
```

```python
#Total number of paid reviews
len(vine_paid_df)
```

```python
#Total number of paid 5-star reviews
len(vine_paid_df[vine_paid_df['star_rating'] == 5.0])
```

```python
#Percentable of paid 5-star reviews
try:
    len(vine_paid_df[vine_paid_df['star_rating'] == 5.0])/len(vine_paid_df)

except:
    print("This results in a Zero Division Error, since the DataFrame is empty.")
```
